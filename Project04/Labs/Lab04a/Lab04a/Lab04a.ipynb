**Question:** What interpretations can you draw from this covariance matrix?

From the heatmap, we can see that the darker or more magenta variables have higher variance. The variables radius mean, peremiter mean, area mean, compactness mean, and concavity mean are variables with higher variance. The other variables seem to have less variance.

```python
cov = cancer_x.cov()
print(cov.shape)

fig, ax = plt.subplots()
im = ax.imshow(cov,cmap = "cool")

ax.set_xticks(np.arange(len(cancer_x.columns)))
ax.set_yticks(np.arange(len(cancer_x.columns)))
ax.set_yticklabels(cancer_x.columns)
ax.set_xticklabels(cancer_x.columns)
ax.set_title("Heatmap of Covariance Data")

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.colorbar(im)
```

## Task 3: Visualizing Principle Component Analysis

### 3a) Principle Component Analysis
1. Use [numpy.linalg.eig](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.eig.html) to compute the eigenvalues and right eigenvectors of the covariance matrix.  
1. Print the eigen_values and the first 2 eigenvectors.
2. Transform the data with the eigen_vectors.
    1. Matrix multiply cancer_x with the first two columns of the eigen_vectors. 
    2. Print the first 5 rows of the transformed data.

**Question:**  Which 2 variables are the first eigenvector most dependent on?  How does that relate to the heatmap?

Concavity mean and concave points mean. Those two variables are higher in variance amongst the other variables and have the darkest/most magenta coloring in the heatmap. 

```python
values,vector = np.linalg.eig(cov)
print("eig values: " + str(values))
print("eig vectors: " + "\n" + str(vector[:,:2]))

mult = cancer_x @ vector[:,:2]
print(mult[:5])
print('''
-----------------------------------
Your results should look like this:
-----------------------------------
Eigen values: 
[0.15632 0.05609 0.01875 0.01012 0.00688 0.00348 0.00001 0.00027 0.00121 0.00206]
First 2 Eigen vectors: 
[[-0.36654  0.33267]
 [-0.1315   0.11754]
 [-0.38051  0.30003]
 [-0.32796  0.28817]
 [-0.16268 -0.33468]
 [-0.34395 -0.32018]
 [-0.43907 -0.17375]
 [-0.47861 -0.04037]
 [-0.16356 -0.34204]
 [-0.05141 -0.58234]]
Transformed data:
           0         1
0 -1.692053 -0.746755
1 -1.105772  0.158411
2 -1.479314 -0.159259
3 -1.335070 -1.286536
4 -1.364975 -0.036868
-----------------------------------
''')
```
